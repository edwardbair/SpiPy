```python
import xarray
import s2cloudless
```

```python
tile = '11SLB'
zarr_store = f'/tablespace/sentinel2/{tile}_sharpend.zarr'
ds = xarray.open_zarr(zarr_store)
```

```python
zarr_store = f'/tablespace/sentinel2/{tile}_60m.zarr'
ds = xarray.open_zarr(zarr_store)
```

```python
ds
```

```python
cloud_detector = s2cloudless.S2PixelCloudDetector(threshold=0.4, average_over=4, dilation_size=2, all_bands=False)
```

```python
reflectances = ds['reflectance'].isel(time=[0,5]).transpose('time', 'y', 'x', 'band').values
```

```python
cloud_prob = cloud_detector.get_cloud_probability_maps(reflectances)
```
