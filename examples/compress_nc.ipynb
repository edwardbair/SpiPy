```python
import xarray
import zarr
import netCDF4
import numpy as np
```

```python
region = 'LBNL'
zarr_name = f'/tablespace/sentinel2/{region}_results.zarr'
ds = xarray.open_zarr(zarr_name)
```

```python
%%time
nc_file = f'/tablespace/sentinel2/{region}_results.nc'
compression_opts = {"zlib": True, "complevel": 5}
ds.to_netcdf(nc_file, mode='w',  encoding={var: compression_opts for var in ds.variables})
```
